```python
import pandas as pd
data = pd.read_csv('Reviews.csv')
data.fillna('')
```

```python
data = data.sort_values(by = 'Time')
time = data[data.columns[7]]
products = data[data.columns[1]]
time = time.values
products = products.values
data.drop('Time', axis = 1, inplace = True)
import numpy as np
time_min = min(time)
time = [x-time_min for x in time]
time = [int(x)/(60*60*24) for x in time]
time = [float(x)/(365.25) for x in time]
time_df = pd.DataFrame(np.array(time).reshape(len(time),1), columns = ['Time'])
```

```python
df_merged =pd.concat([data, time_df], axis=1)
```

```python
df_neutral = df_merged[df_merged['Score'] == 3]
years_neutral = []
year = 0
for i in range(0,12):
    years_neutral.append(len(df_neutral[df_neutral['Time'] <= i+1]) - year)
    year = len(df_neutral[df_neutral['Time'] <= i+1])

years_neutral.append(len(df_neutral[df_neutral['Time'] <= 14]) - year)
```

This metric can help us tell what comments are helpful or non-helpful.
1. If a comment is helpful then the helpful metric value is greater than 1
2. If a comment is not helpful then the helpful metric value is lesser than 1
3. If a comment is equally helpful as is not helpful then the metric is exactly 1
